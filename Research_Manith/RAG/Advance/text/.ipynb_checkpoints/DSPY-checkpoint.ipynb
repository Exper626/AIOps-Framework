```python
import dspy

class RAG(dspy.Signature):
    simple_english = dspy.InputField()
    output = dspy.OutputField()
```

```python
import ollama
lm = dspy.OllamaLocal(model='llama3.2:1b')
```

```python
dspy.settings.configure(lm=lm)
from dspy.signatures.signature import signature_to_template

inputContent = signature_to_template(RAG)
print(str(RAG))
```

```python
class ChainOfThought(dspy.Module):

    def __init__(self);
        super.__init__()
        self.prog = dspy.ChainOfThought(RAG)

    def forward(self,content):
        return self.prog(simple_english = content)

func = ChainOfThought()
func.forward("Prompt")
```
